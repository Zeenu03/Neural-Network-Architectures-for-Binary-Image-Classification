```python
from bing_image_downloader import downloader

downloader.download("koala", limit=100, output_dir='images', force_replace=False, timeout=2,verbose=True)
```

```python
downloader.download("bear", limit=100, output_dir='images', force_replace=False, timeout=2,verbose=True)
```

```python
import os

# Change the file name of the images
for name in ['koala', 'bear']:
    path = f'images/{name}'
    files = os.listdir(path)

    for i,file in enumerate(files):
        os.rename(os.path.join(path, file), os.path.join(path, name + str(i+1) + '.jpg'))
    
    # Check the number of images
    print(len(os.listdir(path)))
```

```python
parent_directory = 'images'
all_images_dir = 'images/all_images'
# Make and write the images to the all_images directory
os.makedirs(all_images_dir, exist_ok=True)
# Copy all images to the "all_images" directory
import os
import shutil


# Iterate over each subfolder in the parent directory
for subdir in os.listdir(parent_directory):
    # Get the full path to the subfolder
    subfolder = os.path.join(parent_directory, subdir)
    
    # Check if the subfolder is a directory
    if os.path.isdir(subfolder):
        # Get the list of files in the subfolder
        files = os.listdir(subfolder)
        
        # Iterate over each file in the subfolder
        for filename in files:
            # Get the full path to the source file
            source_file = os.path.join(subfolder, filename)

            # Get the full path to the destination file
            dest_file = os.path.join(all_images_dir, filename)

            # Copy the file to the "all_images" directory
            shutil.copy2(source_file, dest_file)
```
